```python
import matplotlib.pyplot as plt
import networkx as nx

organism_types = {
    'Trawa': 'Roślina',
    'Zioła': 'Roślina',
    'Liście': 'Roślina',
    'Królik': 'Ssak',
    'Zebra': 'Ssak',
    'Słoń': 'Ssak',
    'Żyrafa': 'Ssak',
    'Koala': 'Ssak',
    'Goryl': 'Ssak',
    'Lis': 'Ssak',
    'Gepard': 'Ssak',
    'Lew': 'Ssak',
    'Lampart': 'Ssak',
    'Jastrząb': 'Ptak',
    'Dingo': 'Ssak',
    'Wilk': 'Ssak',
    'Tygrys': 'Ssak',
    'Orzeł': 'Ptak',
    'Jaguar': 'Ssak',
    'Orangutan': 'Ssak',
    'Sęp': 'Ptak',
    'Pantera': 'Ssak',
    'Harpia': 'Ptak',
    'Puma': 'Ssak',
    'Sokół': 'Ptak',
}

type_colors = {
    'Roślina': '#3CB371',
    'Ssak': '#0088ff',
    'Ptak': '#FFA000',
}

food_chain = {
    'Trawa': ['Królik', 'Zebra'],
    'Zioła': ['Słoń', 'Żyrafa'],
    'Liście': ['Koala', 'Goryl'],
    'Królik': ['Lis', 'Jastrząb'],
    'Zebra': ['Gepard', 'Lew'],
    'Słoń': ['Lew'],
    'Żyrafa': ['Lampart'],
    'Koala': ['Dingo'],
    'Goryl': ['Orangutan'],
    'Lis': ['Wilk', 'Orzeł'],
    'Gepard': ['Lew'],
    'Lew': ['Tygrys'],
    'Lampart': ['Jaguar'],
    'Jastrząb': ['Orzeł'],
    'Dingo': ['Wilk'],
    'Wilk': [],
    'Tygrys': ['Lew'],
    'Orzeł': ['Sęp'],
    'Jaguar': ['Lew'],
    'Orangutan': ['Pantera'],
    'Sęp': ['Harpia'],
    'Pantera': ['Tygrys'],
    'Harpia': ['Puma'],
    'Puma': ['Sokół'],
}

G = nx.DiGraph()

# Dodawanie krawędzi
for predator, prey_list in food_chain.items():
    for prey in prey_list:
        G.add_edge(predator, prey)

# Pozycja węzłów
pos = nx.spring_layout(G, k=20, iterations=10)

# Kolory węzłów na podstawie typu
node_colors = [type_colors[organism_types[node]] for node in G.nodes()]

# Rysowanie grafu
nx.draw_networkx_nodes(G, pos, node_size=500, node_color=node_colors)
nx.draw_networkx_labels(G, pos, font_size=10, font_family='sans-serif')
nx.draw_networkx_edges(G, pos, edge_color='brown', width=2, alpha=0.6, arrows=True)

# Legenda
for label, color in type_colors.items():
    plt.scatter([], [], color=color, label=label, s=100)
plt.legend(scatterpoints=1, frameon=True, labelspacing=1, loc='best')
plt.figtext(0.1, 0.02, "Łańcuch pokarmowy: Rośliny → Roślinożercy → Mięsożercy → Drapieżniki", fontsize=10)
plt.title('Łańcuch Pokarmowy')
plt.axis('off')
plt.show()
```
